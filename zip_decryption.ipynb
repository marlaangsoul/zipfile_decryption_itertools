```python
import itertools
import zipfile

def decryption(password_binding, min, max, zipPath):
    for i in range(min, max+1):
        string = itertools.product(password_binding, repeat=i)
        for j in string:
            password = ''.join(j)
            print(password)
            try:
                ZipPath.extractall(path='./dataset', pwd=password.encode())
                print(f' {password}.')
                print('  !')
                break
            except:
                pass

password_binding = "0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ!@#$%^&*(),.+"
ZipPath = zipfile.ZipFile('./dataset/test.zip')

decryption(password_binding, 1, 3, ZipPath)
```
